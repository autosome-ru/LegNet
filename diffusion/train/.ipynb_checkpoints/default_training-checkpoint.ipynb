```python
CODES = {
    "A": 0,
    "T": 3,
    "G": 1,
    "C": 2,
    'N': 4
}
RIGHT = 'GGTTACGGCTGTT'
LEFT = 'AGTGCTAGCAGGAATGATGCAAAAGGTTCCCGATTCGAACTGCATTTTTTTCACATC'

def n2id(n):
    return CODES[n.upper()]

class Seq2Tensor(nn.Module):
    def __init__(self):
        super().__init__()
    def forward(self, seq):
        if isinstance(seq, torch.FloatTensor):
            return seq
        seq = [n2id(x) for x in seq]
        code = torch.from_numpy(np.array(seq))
        code = F.one_hot(code, num_classes=5)
        
        code[code[:, 4] == 1] = 0.25
        code = code[:, :4].float()
        return code.transpose(0, 1)
s2t = Seq2Tensor()
right_s2t = torch.concat((s2t(RIGHT), torch.zeros(2, s2t(RIGHT).shape[1])))
left_s2t = torch.concat((s2t(LEFT), torch.zeros(2, s2t(LEFT).shape[1])))
```

```python
def initialize_weights(m):
    if isinstance(m, nn.Conv1d):
        n = m.kernel_size[0] * m.out_channels
        m.weight.data.normal_(0, math.sqrt(2 / n))
        if m.bias is not None:
            nn.init.constant_(m.bias.data, 0)
    elif isinstance(m, nn.BatchNorm1d):
        nn.init.constant_(m.weight.data, 1)
        nn.init.constant_(m.bias.data, 0)
    elif isinstance(m, nn.Linear):
        m.weight.data.normal_(0, 0.001)
        if m.bias is not None:
            nn.init.constant_(m.bias.data, 0)
            
class DataloaderWrapper:
    def __init__(self, dataloader, batch_per_epoch):
        self.batch_per_epoch = batch_per_epoch
        self.dataloader = dataloader
        self.iterator = iter(dataloader)

    def __len__(self):
        return self.batch_per_epoch
    
    def __next__(self):
        try:
            return next(self.iterator)
        except StopIteration:
            self.iterator = iter(self.dataloader)

    def __iter__(self):
        for _ in range(self.batch_per_epoch):
            try:
                yield next(self.iterator)
            except StopIteration:
                self.iterator = iter(self.dataloader)
```

```python
class Trainer:
    def __init__(self,
            model: torch.nn.Module, 
            pretraned_model: torch.nn.Module,
            train_dataloader: torch.utils.data.DataLoader ,
            test_dataloader: torch.utils.data.DataLoader ,
            criterion: torch.nn.CrossEntropyLoss,
            loss_criterion: torch.nn.CrossEntropyLoss,
            optimizer: torch.optim.Optimizer,
            epochs: int,
            N: int,
            scheduler: torch.optim.lr_scheduler.OneCycleLR = None,
            batch_size: int = 1024,
            batch_per_epoch: int = 1000,
            device = torch.device("cuda:0"),
            val: bool = True,
            time = 0
            ):
        self.optimizer = optimizer
        self.criterion = criterion
        self.score_criterion = loss_criterion
        self.model = model
        self.pretraned_model = pretraned_model
        self.train_dl = train_dataloader
        self.test_dl = test_dataloader
        self.epochs = epochs
        self.batch_per_epoch = batch_per_epoch
        self.N = N
        self.device = device
        self.scheduler = scheduler
        self.batch_size = batch_size
        self.val = val
        self.time = time
        self.score_cor_mean = []
        self.score = []
            
    def train(self, epoch):
        print(f'start training, epoch = {epoch}')
        self.model.train()
        ltr = []
        for b_i, data in tqdm(enumerate(self.train_dl), mininterval=60):
            target_seq, mutated_seq, mut_pos = data
            target_seq, mutated_seq = target_seq.float().to(self.device), mutated_seq.float().to(self.device) 
            target_score = mutated_seq[:,4,1].clone()
            pred = self.model(mutated_seq)
            loss = self.criterion(pred, target_seq)
            ltr.append(loss.item())
            loss.backward()
            self.optimizer.step()
            if self.scheduler:
                self.scheduler.step()
            self.optimizer.zero_grad() 
            
#             if b_i % (self.batch_per_epoch) == 0:
#                 self.type = 'train'
#                 self.seqsplot(mut_pos, epoch, target_seq, mutated_seq)
        mean_loss = np.mean(ltr)
        return mean_loss
    
    def validate(self, epoch):
        print(f'start validating, epoch = {epoch}')
        with torch.no_grad():
            self.model.eval()
            self.pretraned_model.eval()
            lte = []
            score_losses = []
            score_cores = []
            
            
            for b_i, data in tqdm(enumerate(self.test_dl), mininterval=60):
                target_seq_val, mutated_seq_val, mut_pos_val = data
                mutated_seq_val, target_seq_val = mutated_seq_val.float().to(self.device), target_seq_val.float().to(self.device)
                target_score = mutated_seq_val[:,4,1].clone()
                mutated_seq = mutated_seq_val[:,:4,:].clone()
                mut_seqs = torch.concat((mutated_seq, torch.zeros(mutated_seq.shape[0],2,mutated_seq.shape[2], device=self.device)), dim=1)
                left_batch = torch.broadcast_to(left_s2t, (mut_seqs.shape[0], left_s2t.shape[0], left_s2t.shape[1])).to(self.device)
                right_batch = torch.broadcast_to(right_s2t, (mut_seqs.shape[0], right_s2t.shape[0], right_s2t.shape[1])).to(self.device)
                long_mut_seqs = torch.concat((left_batch, mut_seqs, right_batch), dim=2).to(self.device)
                mut_score = self.pretraned_model(long_mut_seqs)[1]
                pred = self.model(mutated_seq_val)
                loss = self.criterion(pred, target_seq_val)
                lte.append(loss.item())
                pred_seq = torch.softmax(pred, dim=1)
                seqs = torch.concat((pred_seq, torch.zeros(pred_seq.shape[0],2,pred_seq.shape[2], device=device)), dim=1)
                
                long_pred = torch.concat((left_batch, seqs, right_batch), dim=2)
                pred_score = self.pretraned_model(long_pred)[1]
                
                score_loss = self.score_criterion(pred_score, target_score)
                score_losses.append(score_loss.item())
                score_cor = stats.pearsonr(pred_score.cpu().numpy(), target_score.cpu().numpy())[0]
                score_cores.append(score_cor)

                if b_i % (self.batch_per_epoch) == 0:
                    self.seqsplot(mut_pos_val[0], epoch, target_seq_val[0], mutated_seq_val[0][:4], pred_seq[0], pred_score[0], target_score[0], mut_score[0])
                
            self.score.append(np.mean(score_losses))
            self.score_cor_mean.append(np.mean(score_cores))
            mean_loss_val = np.mean(lte)
            return mean_loss_val
       
        
    def training(self):
        
        self.save_dir = f"/home/kekulen/dna-ddpm/data/training/saved_model/time_{self.time}_scheduler_{type(self.scheduler).__name__}_epochs_{self.epochs}"
        os.makedirs(self.save_dir, exist_ok=True)
        train_losses = []
        test_losses = []
        for epoch in tqdm(range(self.epochs)):
            tr_loss = self.train(epoch)
            train_losses.append(tr_loss)
            
            if self.val is True:
                test_loss = self.validate(epoch)
                test_losses.append(test_loss)
            
            if epoch % 3 == 0:
                clear_output()
            self.plotter(train_losses,test_losses, epoch)
            self.save_model(epoch,train_losses)
        return train_losses, test_losses, self.score
    
    def seqsplot(self, mutate_pos, epoch, target_to_pic, mutated_seq, pred_seq, pred_score_to_pic, real_score_to_pic, mut_score_to_pic):
        fig, (ax1, ax2, ax3) =  plt.subplots(3, 1,  figsize=(14, 4), sharex=True)

        tracer = seqencoder.MotifsTracer(motifs=["A"],exact=False, inexact=False)
        tracer.visualize_similarity(target_to_pic.cpu().numpy(), ax=ax1)
        ax1.set_title(f'Target, expr = {real_score_to_pic:.04f}', loc='left', fontsize='medium')
        mp = np.where(mutate_pos)
        ax1.plot(mp, 0, marker='+', color="red")
        ax2.plot(mp, 0, marker='+', color="red")
        ax3.plot(mp, 0, marker='+', color="red")

        ax2.set_title(f'Reconstructed, expr={pred_score_to_pic.item():.04f}',  loc='left', fontsize='medium')
        
        tracer.visualize_similarity(pred_seq.cpu().numpy(), ax=ax2)
        
        tracer.visualize_similarity(mutated_seq.cpu().numpy(), ax=ax3)
        ax3.set_title(f'Mutated, expr = {mut_score_to_pic.item():.04f}', loc='left', fontsize='medium')

        subtitle_string = f'Epoch={epoch}'
        plt.suptitle(subtitle_string, fontsize=18)
        pic_seq_name = os.path.join(self.save_dir, f'seq_pic{epoch}.png')
        plt.tight_layout()
        fig.savefig(pic_seq_name)
        plt.show()    
    

    def plotter(self, loss_train, loss_val, epoch):
        fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5, 1,  figsize=(7, 7))
        
        ax1.plot(loss_train, color='red')
        ax3.plot(loss_train, color='red')
        ax2.plot(loss_val, color='blue')
        ax3.plot(loss_val, color='blue')
        ax4.plot(self.score, color = 'black')
        ax5.plot(self.score_cor_mean, color = 'black')
        ax1.grid(axis='x')
        ax2.grid(axis='x')
        ax3.grid(axis='x')
        ax2.set_xlabel('Epoch')
        ax1.set_ylabel('Train Loss')
        ax3.set_ylabel('Train and val Loss')
        ax2.set_ylabel('Val Loss')
        ax4.set_ylabel('Score MSE')
        ax5.set_ylabel('Pearson cor')
        
        suptitle_string = f'epoch={epoch}'
        fig.suptitle(suptitle_string, y=1.05, fontsize=10)

        pic_test_name = os.path.join(self.save_dir, f"lossestrainandtest_epoch={epoch}.png")
        plt.tight_layout()
        fig.savefig(pic_test_name)
        fig.show()
        with open(f'/home/kekulen/dna-ddpm/data/training/saved_model/time_{self.time}_scheduler_{type(self.scheduler).__name__}_epochs_{self.epochs}/train_loss.npy', 'wb') as train_loss_f:
            np.save(train_loss_f, np.array(loss_train))
        with open(f'/home/kekulen/dna-ddpm/data/training/saved_model/time_{self.time}_scheduler_{type(self.scheduler).__name__}_epochs_{self.epochs}/test_loss.npy', 'wb') as test_loss_f:
            np.save(test_loss_f, np.array(loss_val))
        with open(f'/home/kekulen/dna-ddpm/data/training/saved_model/time_{self.time}_scheduler_{type(self.scheduler).__name__}_epochs_{self.epochs}/score_loss.npy', 'wb') as score_loss_f:
            np.save(score_loss_f, np.array(self.score))
        with open(f'/home/kekulen/dna-ddpm/data/training/saved_model/time_{self.time}_scheduler_{type(self.scheduler).__name__}_epochs_{self.epochs}/score_loss.npy', 'wb') as score_cor_f:
            np.save(score_cor_f, np.array(self.score_cor_mean))
            
    def save_model(self, epoch, losseshist):
        PATH = os.path.join(self.save_dir, f"model_{epoch}.pth")
            
        torch.save({
            'epoch' : epoch,
            'model_state_dict': self.model.state_dict(),
            'optimizer_state_dict': self.optimizer.state_dict(),
            'loss': losseshist
            }, PATH)

        print(f'---------------  SAVED MODEL {PATH}-------------------')
```

```python
time_n = time.strftime('%d_%b_%Y_%H-%M')
train = Trainer(
            model=legnet, 
            pretraned_model=pretraned_model,
            train_dataloader=dl_train,
            test_dataloader=dl_test,
            criterion=criterion,
            loss_criterion=score_criterion,
            optimizer=optimizer,
            epochs=epochs,
            N=N,
    
            scheduler=scheduler,
            batch_size=batch_size,
            batch_per_epoch=batch_per_epoch,
            device = device,
            val = True,
            time = time_n
            )
try:
    train_losses, test_losses, weights = train.training()
except Exception as e:
    logger.exception("Training failed")
```
